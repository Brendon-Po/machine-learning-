```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn import svm
from sklearn import metrics
```

```python
df = pd.read_csv('../SVM/data.csv')
df.head()
```

```python
target = df[['target']]
df = df.drop('target', axis=1)
df = df.drop('song_title', axis=1)
df = df.drop('id', axis=1)
```

## Label Encoder ##

```python
from sklearn.preprocessing import LabelEncoder
#agora queremos aplicar o fit_transform que 'treina' o encoder para a coluna que queremos e transforma os valores
#textuais em numéricos

enc = LabelEncoder()
inteiros = enc.fit_transform(df['artist'])
len(set(inteiros)) #1343 artistas diferentes
```

```python
df['artist_num'] = inteiros
df.drop('artist', axis=1, inplace=True)
df['artist_num']
```

```python
df['artist_num'].unique()
df
```

## One Hot Encoding (OHE) ##

```python
ohe = OneHotEncoder()
df.values
```

```python
new_art = ohe.fit_transform(df[['artist']])
new_art.toarray() #transforma em array denso porque o padrão é sparse matrix e é difícil de visualizar.

df = df.drop('artist', axis=1)
df= pd.concat([df, pd.DataFrame(new_art.toarray())], axis=1)
df
```

## Pandas Get_Dummies ##

```python
df
```

```python
df_dummies = pd.get_dummies(df, columns = ['artist'], prefix = 'artist')
df_dummies
```

## Pipelines ##


```python
from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.model_selection import cross_val_predict
```

```python
target
```

```python
clf = svm.SVC().fit(df_dummies, target.values.ravel())
```

```python
target.values.ravel()
```

```python
clf
```

```python
def acuracia(clf, df, target):
    resultados = cross_val_predict(clf, df, target, cv = 10 )
    return metrics.accuracy_score(target, resultados)
```

```python
acuracia(clf, df_dummies, target.values.ravel())
```

```python
#fazendo o pipeline
pip1 = Pipeline([ ('scaler', StandardScaler()), ('svm', svm.SVC())])
pip1

pip2 = Pipeline( [ ('scaler', MinMaxScaler()), ('svm', svm.SVC(kernel = 'rbf'))])
```
